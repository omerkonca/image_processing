```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
img1 =cv2.imread("m1.png",cv2.IMREAD_COLOR)
temp=0
for i in range(img1.shape[0]):
    for j in range (img1.shape[1]):
        temp=img1[i][j][0]
        img1[i][j][0]=img1[i][j][2]
        img1[i][j][2]=temp
#print(img1)
plt.imshow(img1)
```

```python
img2 =cv2.imread("m2.png",cv2.IMREAD_COLOR)
temp=0
for i in range(img2.shape[0]):
    for j in range (img2.shape[1]):
        temp=img2[i][j][0]
        img2[i][j][0]=img2[i][j][2]
        img2[i][j][2]=temp
#print(img2)
plt.imshow(img2)
```

```python
genislik1=(img1.shape[0])
yukseklik1=(img1.shape[1])
#print(genislik1)
#print(yukseklik1)
```

```python
sayac=0
fark = cv2.absdiff(img1,img2)
for i in range(fark.shape[0]):
    for j in range (fark.shape[1]):
         fark[i][j][0]!=fark[0][0][0]
         sayac+=1
#print(fark)

plt.imshow(fark)
print("toplam deien pixel adedi oran :", sayac*100 / (genislik1 * yukseklik1 * 24))
```
